```python
Ages = df[df.columns[df.columns.str.contains('age')].tolist()]
```

```python
Ages
```

```python
def extract_wave_number(col_name):
    # Assuming the format is 'wXXage', where XX is the wave number
    return int(col_name[1:3])
```

```python
AgesCopy  = Ages.copy()
```

```python
for col1 in Ages.columns:
    wave_number_col1 = extract_wave_number(col1)

    # Calculate the mean differences for the current column
    mean_differences = means - means[col1]

    # Fill missing values in the current column
    for col2 in Ages.columns:
        #wave_number_col2 = extract_wave_number(col2)

        # Ensure col2 is from a later wave
        #if wave_number_col2 > wave_number_col1:
            # Use original, non-imputed data from later waves
        Ages.loc[Ages[col1].isnull() & AgesCopy[col2].notnull(), col1] = AgesCopy[col2] - mean_differences[col2]
```
